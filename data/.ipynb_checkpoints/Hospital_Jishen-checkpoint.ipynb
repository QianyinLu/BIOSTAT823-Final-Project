```python
import json
from urllib.request import Request, urlopen

import pandas as pd
```

```python
url1 = "https://healthdata.gov/api/action/datastore/search.json?resource_id=38e07d07-40a2-4749-bfd3-68fefa42a61f&limit=53"
request1 = Request(url1)
```

```python
html1 = urlopen(request1)
data1 = html1.read()
```

```python
strs1 = str(data1)
strs_for_json1 = strs1[136:-2]
datas1 = json.dumps(strs_for_json1)
data_json1 = json.loads(strs_for_json1)
df1 = pd.DataFrame(data_json1['records'])
```

```python
df1 = df1.drop(["reporting_cutoff_start", "entry_id"], axis=1)
```

```python
df1.to_csv("data/hospital1.csv")
```

```python
origin2 = pd.read_csv("https://www.cdc.gov/nhsn/pdfs/covid19/covid19-NatEst.csv")
```

```python
df2 = origin2.drop(labels=0, axis=0).drop(labels="Notes", axis=1)
df2["day"] = df2.collectionDate.str[:2].astype(int)
df2["month"] = df2.collectionDate.str[2:5]
df2["year"] = df2.collectionDate.str[5:].astype(int)
df2 = df2.drop("collectionDate", axis=1)
```
